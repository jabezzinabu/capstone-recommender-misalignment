```python
import pandas as pd
import gzip
import json
import os

# First confirm file exists
os.listdir("../data/external")
```

```python
path = "../data/external/Video_Games_5.json.gz"  # adjust name if slightly different

def parse_gz(path):
    with gzip.open(path, "rt", encoding="utf-8") as f:
        for line in f:
            yield json.loads(line)

df = pd.DataFrame(parse_gz(path))

df.head()
```

```python
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nMissing values:\n", df.isnull().sum().sort_values(ascending=False).head(10))
print("\nRating distribution:\n", df["overall"].value_counts().sort_index())
```

```python
sample = df.sample(5000, random_state=42)
sample.to_csv("../data/raw/amazon_videogames_sample.csv", index=False)
print("Saved sample")
```
