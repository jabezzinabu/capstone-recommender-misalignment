```python
import pandas as pd

# Path to ratings file
file_path = "../data/external/ml-1m/ml-1m/ratings.dat"

ratings = pd.read_csv(
    file_path,
    sep="::",
    engine="python",
    names=["user_id", "movie_id", "rating", "timestamp"]
)

ratings.head()
```

```python
print("Shape:", ratings.shape)
print("Missing values:\n", ratings.isnull().sum())
print("Unique users:", ratings["user_id"].nunique())
print("Unique movies:", ratings["movie_id"].nunique())
print("Rating distribution:\n", ratings["rating"].value_counts().sort_index())
```
